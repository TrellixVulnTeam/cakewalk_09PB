* ## Softmax function
[wiki](https://en.wikipedia.org/wiki/Softmax_function)
[維基百科](https://zh.wikipedia.org/wiki/Softmax%E5%87%BD%E6%95%B0)

<img style="float: left;" src="https://wikimedia.org/api/rest_v1/media/math/render/svg/e348290cf48ddbb6e9a6ef4e39363568b67c09d3">


```python
z = [1.0, 2.0, 3.0, 4.0, 1.0, 2.0, 3.0]
z_exp = np.exp(z) # [math.exp(i) for i in z]
print(z_exp)
```

```python
sum_z_exp = sum(z_exp)

softmax = np.around((z_exp/sum_z_exp), decimals=3) # [round(i / sum_z_exp, 3) for i in z_exp]
print(softmax)
```

The softmax of that is [0.024, 0.064, 0.175, 0.475, 0.024, 0.064, 0.175]. The output has most of its weight where the '4' was in the original input. This is what the function is normally used for: <span style="color:red; font-size: 1.2em;">**to highlight the largest values and suppress values which are significantly below the maximum value**</span>. <br />

- ## Sigmoid
[wiki](https://en.wikipedia.org/wiki/Sigmoid_function)

<img style="float: left;" src="https://wikimedia.org/api/rest_v1/media/math/render/svg/9537e778e229470d85a68ee0b099c08298a1a3f6"></img>



```python
X = np.arange(-10, 11, 0.2, dtype=float)
S = 1 / (1 + np.exp(-X))

plt.figure(figsize=(8,5))
plt.plot(X, S)
plt.grid(True)

plt.xlabel('x')
plt.ylabel('S(x)')

plt.xlim(-10.5, 10.5)
plt.ylim(-0.05, 1.05)
y_ticks = np.arange(0., 1.1, 0.1)
plt.yticks(y_ticks)

pX = np.arange(-3, 4, 1)
pS = 1 / (1 + np.exp(-pX))
print(pS)
for x, s in zip(pX, pS):
    atxt = '(%d,  %.3f)' % (int(x), round(s,3))
    plt.plot(x, s, 'o')
    plt.annotate(atxt, xy=(x, s), xytext=(x+0.5, s))
plt.show()
```

1. X = 0, S = 1 / (1 + 1) = 0.5
2. X < -3 and X > 3, S has been very close to 0 and 1 respectively.
3. <span style="color:red; font-size: 1.2em;">**Hard Sigmode**</span> : σ(x) = clip((x + 1)/2, 0, 1) = max(0, min(1, (x + 1)/2))
